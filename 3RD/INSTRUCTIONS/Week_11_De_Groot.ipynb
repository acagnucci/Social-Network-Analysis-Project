# DeGroot Influence

We now consider a degroot model where agent listens uniformly to their neighbors (successors) with a total weight of 1/2 and to themselves with a weight 1/2 at every stage. 

We first construct the matrix associated to te graph and the previous story. We can then create a program running the simulation using matrix multiplication between matrices to compute the vector of opinions at every stage and draw some pictures.

We will then compute numerically the influence of every agent by looking at the final average opinion starting from a value of 1 for a given agent and 0 for the other. Notice that the graph is not connected hence there is not eòergence of consensus a priori.

## Contructing of the matrix

We first construct the matrix where each agent listen uniformly on his neighbors. To check that the computation that we are doing are correct, we will focus on the cose of 'Kmoch'.

```python
def stochastique(A): # normaliwe the matrix
    (n,m)=A.shape
    B=np.zeros((n,m)) # create the new matrix
    s=A.sum(axis=1) # sum on the line
    
    for i in range(n):
        if s[i]==0: # problematic case, we make this agent put some weight on itself
            B[i,i]=1
        else:
            B[i,:]=A[i,:]/s[i]
    
    return B
```

```python
T1=stochastique(B1)
```

## Simulation (fixed number of iteration)

```python
def simulation(T,p,L): # matrix, initial vector, number of iteration
    
    pc=p  # the vector of opinion at the current stage
    numb=T.shape[1]
    
    for n in range(L):
        if n==0: # the first vector is the initial one
            data_p=p
        else:
            data_p=np.concatenate((data_p,pc),axis=1)
            # concatenate the new vector to the previous one
    
        pc=np.matmul(T,pc)# compute the new vector of opinions by doing the multiplication
    
    
    data_x=range(L) # create a list of the number of iteration
    
    for i in range(numb):
        plt.plot(data_x,data_p[i,:]) # plot for each agent i, i plot the evolution of the opinion of this agent i across time
        
    plt.show()
    
    return pc.sum()/numb # return the average opinion in the network

```

```python
maxi=10
p=np.zeros((N_SNA,1))
p[2]=5
simulation(T,p,maxi)
```

# Comparison of the centralities

Who should the menza offer a meal in exchange of a good grade? To do so, we will investigate the consequence of each agent putin a grade of 5 when all the other agents put a grade of 0.

```python
def comparisondegroot(n,T,iteration): # number of node, size of the matrix, number of iteration 
    data=[] # data to save our result
    numb=T.shape[1]
    
    for node in range(n):
        x=0
        p0=np.zeros((numb,1))
        p0[node]=5 # initial condition where only one guy put a 5 and all the other 0
        
        y=simulation(T,p0,iteration)
        
        data.append(y)
        
    return data
```

# Endogenize stopping time

```python
def simulation_endo(T,p,L): # matrix, initial vector, number of iteration
    
    po=p
    pc=p  # the vector of opinion at the current stage
    numb=T.shape[1]
    
    count=0
    error=100001
    while (error>=0.0001) and (count<=L):
        if count==0: # the first vector is the initial one
            data_p=p
        else:
            data_p=np.concatenate((data_p,pc),axis=1)
            # concatenate the new vector to the previous one
        
        po=pc
        
        pc=np.matmul(T,pc)# compute the new vector of opinions by doing the multiplication
        count=count+1
        
        error=0
        for i in range(numb):
            error=error+abs(po[i]-pc[i])
        
    
    data_x=range(count) # create a list of the number of iteration
    
    for i in range(numb):
        plt.plot(data_x,data_p[i,:]) # plot for each agent i, i plot the evolution of the opinion of this agent i across time
        
    plt.show()
    
    return pc.sum()/numb # return the average opinion in the network

```
