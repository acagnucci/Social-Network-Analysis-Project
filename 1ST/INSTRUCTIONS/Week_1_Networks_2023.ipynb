# 5) Creating a network edge by edge from another data structure

From an alternative data structure, there are two options:
- create first the list of edges or the adjacency matrix and use the previous section,
- directly go over the list of the data structure and add edges progressively.

We present two functions doing that:
- the first one takes as input a matrix and create the graph (we are writing our own nx.from_numpy_matrix).
- the second one takes as input an excel file where each line has two entries representing the two extremities of a link.

The idea is to parse the input (matrix or excel file) and convert each entry (coefficient of the matrix, line in the excel file) into an edge.

What do we need? Loop and add_edge.

## From matrix

```python
def digraph_from_adjacency_matrix(mat): # define a new complex function taking an argument.
    # The argument will be called mat in the program independtly of the name outside. 
    # Mat ONLY exists inside the function
    
    G=nx.DiGraph()
    
    n=mat.shape[1]
    
    for i in range(n):
        for j in range(n):
            if mat[i,j]==1:
                G.add_edge(i,j)

    return G  # function return the graph G, G is also an internal name
```
